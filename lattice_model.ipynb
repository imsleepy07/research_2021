```python
import matplotlib as mpl
import numpy as np
import networkx as nx
import scipy
import scipy.stats
import math
import matplotlib.pyplot as plt

import random
```

```python
def g_process_degreeSequencePDF(degreeSequence):
    max_degree = max(degreeSequence)
    retx, rety = [], []
    normalization_condition = len(degreeSequence)
    for k in range(max_degree + 1):
        count = 0
        for l in degreeSequence:
            if l == k:
                count += 1
        retx.append(k)
        rety.append(count / normalization_condition)
    return retx, rety
```

```python
root = nx.triangular_lattice_graph(10, 10)
nx.draw(root)
```

```python
degs = root.degree()
deg_seq = [d for n, d in root.degree()]

p = 0.2
to_delete = []
for node in root.nodes():
    k = degs[node]
    P = random.uniform(0, 1)
    if P <= p / k:
        to_delete.append(node)
for node in to_delete:
    root.remove_node(node)

nx.draw(root)
```

```python
deg_seq = [d for n, d in root.degree()]
x, y = g_process_degreeSequencePDF(deg_seq)
plt.plot(x, y)
```
